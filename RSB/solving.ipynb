```julia
Φ(x) = erfc(-x/sqrt(2))/2;
w1(x) = exp(-x^2/2)/sqrt(2π) + x*Φ(x);
w2(x) = x*exp(-x^2/2)/sqrt(2π) + (1+x^2)*Φ(x);
```

```julia
g(eR,μc,m) = μc*eR - m;
κ(eN,μe,K,γ) = K - μe*eN/γ;
σg(qR,σc,σm) = sqrt(σm^2 + σc^2*qR);
σκ(qN,σe,σK,γ) = sqrt(σK^2 + σe^2*qN/γ);
Δκ(eN,qN,μe,σe,K,σK,γ) = κ(eN,μe,K,γ)/σκ(qN,σe,σK,γ);
Δg(eR,qR,μc,σc,m,σm) = g(eR,μc,m)/σg(qR,σc,σm);
# g(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = μc*eR - m;
# κ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = K - μe*eN/γ;
# σg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = sqrt(σm^2 + σc^2*qR);
# σκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = sqrt(σK^2 + σe^2*qN/γ);
# Δκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = κ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/σκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ);
# Δg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = g(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/σg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ);
```

```julia
ϕNeq(ϕN,eR,qR,μc,σc,m,σm) = ϕN - (Φ(Δg(eR,qR,μc,σc,m,σm)));
ϕReq(ϕR,eN,qN,μe,σe,K,σK,γ) = ϕR - (Φ(Δκ(eN,qN,μe,σe,K,σK,γ)));
νeq(ϕN,ν,χ,σc,σe,K,σK,m,σm) = ν - (
    -ϕN/(σc*σe*ρ*χ)
);
χeq(ϕR,ν,χ,σc,σe,γ,ρ) = χ - (
    ϕR/(1-ρ*σe*σc*ν/γ)
);
eNeq(χ,eN,eR,qR,μc,σc,σe,m,σm,ρ) = eN - (
    (σg(qR,σc,σm)/(σc*σe*ρ*χ))*w1(Δg(eR,qR,μc,σc,m,σm))
);
eReq(ν,eN,eR,qN,σc,μe,σe,K,σK,γ,ρ) = eR - (
    (σκ(qN,σe,σK,γ)/(1-ρ*σe*σc*ν/γ))*w1(Δκ(eN,qN,μe,σe,K,σK,γ))
);
qNeq(χ,eR,qN,qR,μc,σc,σe,m,σm,ρ) = qN - (
    (σg(qR,σc,σm)/(σc*σe*ρ*χ))^2*w2(Δg(eR,qR,μc,σc,m,σm))
);
qReq(ν,eN,qN,qR,σc,μe,σe,K,σK,γ,ρ) = qR - (
    (σκ(qN,σe,σK,γ)/(1-ρ*σe*σc*ν/γ))^2*w2(Δκ(eN,qN,μe,σe,K,σK,γ))
);
RSBcrit(ϕR,χ,ρ) = ρ^2 * ϕR*(2χ-ϕR) - 1;
# ϕNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = ϕN - (Φ(Δg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)));
# ϕReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = ϕR - (Φ(Δκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)));
# νeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = ν - (
#     -ϕN/(σc*σe*ρ*χ)
# );
# χeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = χ - (
#     ϕR/(1-ρ*σe*σc*ν/γ)
# );
# eNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = eN - (
#     (σg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/(σc*σe*ρ*χ))*w1(Δg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ))
# );
# eReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = eR - (
#     (σκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/(1-ρ*σe*σc*ν/γ))*w1(Δκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ))
# );
# qNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = qN - (
#     (σg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/(σc*σe*ρ*χ))^2*w2(Δg(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ))
# );
# qReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = qR - (
#     (σκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)/(1-ρ*σe*σc*ν/γ))^2*w2(Δκ(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ))
# );
# RSBcrit(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = ρ^2 * ϕR*(2χ-ϕR) - 1;
```

```julia
lsqobj(μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = (ϕN,ϕR,ν,χ,eN,eR,qN,qR) -> (
    ϕNeq(ϕN,eR,qR,μc,σc,m,σm)^2
    +
    ϕReq(ϕR,eN,qN,μe,σe,K,σK,γ)^2
    +
    νeq(ϕN,ν,χ,σc,σe,K,σK,m,σm)^2
    +
    χeq(ϕR,ν,χ,σc,σe,γ,ρ)^2
    +
    eNeq(χ,eN,eR,qR,μc,σc,σe,m,σm,ρ)^2
    +
    eReq(ν,eN,eR,qN,σc,μe,σe,K,σK,γ,ρ)^2
    +
    qNeq(χ,eR,qN,qR,μc,σc,σe,m,σm,ρ)^2
    +
    qReq(ν,eN,qN,qR,σc,μe,σe,K,σK,γ,ρ)^2
);
# lsqobj(μc,σc,μe,σe,K,σK,m,σm,γ,ρ) = (ϕN,ϕR,ν,χ,eN,eR,qN,qR) -> (
#     ϕNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     ϕReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     νeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     χeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     eNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     eReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     qNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
#     +
#     qReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
# );
```

```julia
crit_lsqobj(μc,σc,μe,σe,K,σK,m,σm,γ) = (ϕN,ϕR,ν,χ,eN,eR,qN,qR,ρ) -> (
    ϕNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    ϕReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    νeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    χeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    eNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    eReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    qNeq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    qReq(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
    +
    RSBcrit(ϕN,ϕR,ν,χ,eN,eR,qN,qR,μc,σc,μe,σe,K,σK,m,σm,γ,ρ)^2
);
```

```julia
varps(ρ) = [1e0M, 2e-1sqrt(M), 1e0M, 2e-1sqrt(M), 1e0, 1e-1, 1e0, 1e-1, M/S, ρ];
```

```julia
function solve_consist(ps,ρ)
    optf = OptimizationFunction((x,p)->lsqobj(p...)(x...));
    u0 = 0.1rand(8);
    # (ϕN,ϕR,ν,χ,eN,eR,qN,qR)
    prob = OptimizationProblem(optf,u0,ps(ρ),lb=vcat(zeros(2),-1e1ones(2),zeros(2),zeros(2)),ub=vcat(ones(2),1e1ones(2),1e1ones(2),1e1ones(2)));
    # sol = solve(prob,BBO_adaptive_de_rand_1_bin_radiuslimited(), maxiters=1_000_000,maxtime=5)#,abstol=1e-14,local_abstol=1e-14,local_reltol=1e-14)
    # sol = solve(prob,BBO_de_rand_2_bin(), maxiters=1_000_000,maxtime=5)#,abstol=1e-14,local_abstol=1e-14,local_reltol=1e-14)
    sol = solve(prob,BBO_adaptive_de_rand_1_bin(), maxiters=1_000_000,maxtime=5)#,abstol=1e-14,local_abstol=1e-14,local_reltol=1e-14)
    return sol.u
end
```
